```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/2019.csv")
```

```python
data.info()
```

```python
data.isnull().sum().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of Happiness Score
sns.histplot(data['Happiness Score'], bins=10, kde=True)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

# Boxplot for GDP per Capita
sns.boxplot(y='GDP per capita', data=data)
plt.title('Distribution of GDP per Capita')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot between Happiness Score and GDP per Capita
sns.scatterplot(x='GDP per capita', y='Happiness Score', data=data)
plt.title('Happiness Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
# Select only numerical features for correlation calculation
numerical_data = data.select_dtypes(include=['number'])  # This line is added to filter numerical columns
sns.heatmap(numerical_data.corr(), annot=True, cmap="coolwarm")  # Updated to use numerical_data
plt.title('Correlation between Happiness Factors')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Top 10 happiest countries
top_countries = data.nlargest(10, 'Happiness Score')
sns.barplot(x='Happiness Score', y='Country or region', data=top_countries)
plt.title('Top 10 Happiest Countries')
plt.xlabel('Happiness Score')
plt.show()

# Bottom 10 least happy countries
bottom_countries = data.nsmallest(10, 'Happiness Score')
sns.barplot(x='Happiness Score', y='Country or region', data=bottom_countries)
plt.title('Bottom 10 Least Happy Countries')
plt.xlabel('Happiness Score')
plt.show()
```

```python
# Scatter plot for Happiness Score vs. GDP per Capita
sns.scatterplot(x='GDP per capita', y='Happiness Score', data=data)
plt.title('Happiness Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Scatter plot for Happiness Score vs. Social Support
sns.scatterplot(x='Social support', y='Happiness Score', data=data)
plt.title('Happiness Score vs Social Support')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.show()
```
